# Podstawowe pojęcia z zakresu elektryczności

## 1. Podstawowe pojęcia

- **Napięcie elektryczne (U)**: różnica potencjałów elektrycznych między dwoma punktami obwodu. Jednostka: **wolt (V)**  
  Przykład: napięcie w gniazdku domowym wynosi 230 V

- **Opór elektryczny (R)**: opór, jaki stawia materiał przepływowi prądu. Jednostka: **om (Ω)**  
  Przykład: opór żarówki 40 W wynosi około 1320 Ω

- **Natężenie prądu elektrycznego (I)**: ilość ładunków elektrycznych przepływających przez przekrój przewodnika w jednostce czasu. Jednostka: **amper (A)**  
  Przykład: prąd płynący przez czajnik elektryczny wynosi około 8,5 A

## 2. Wzory podstawowe

- **Prawo Ohma**:  
  $$U = I \cdot R$$
- **Obliczanie oporu**:  
  $$R = \frac{U}{I}$$
- **Obliczanie natężenia prądu**:  
  $$I = \frac{U}{R}$$

## 3. Obliczanie mocy

- **Moc elektryczna**:  
  $$P = U \cdot I$$
- **Moc z oporu**:  
  $$P = \frac{U^2}{R}$$
- **Moc z natężenia**:  
  $$P = I^2 \cdot R$$

## 4. Bezpieczeństwo użytkowania urządzeń elektrycznych

- **Dobór napięcia**: Urządzenia elektryczne są projektowane do pracy przy określonym napięciu  
  - Podłączenie do niższego napięcia → urządzenie działa słabiej lub wcale  
  - Podłączenie do wyższego napięcia → urządzenie może ulec uszkodzeniu

- **Bezpieczniki**: chronią obwody przed przeciążeniem i zwarciem.  
  W domowych instalacjach stosuje się bezpieczniki o wartościach np. 10 A lub 16 A

## 1. Prawo Ohma
Żarówka ma opór 120 Ω. Oblicz natężenie prądu płynącego przez żarówkę, jeśli podłączymy ją do sieci o napięciu 230 V.

**Rozwiązanie:**
$$
I = \frac{U}{R} = \frac{230}{120} \approx 1,92 \text{ A}
$$

## 2. Moc z oporu
Grzałka żelazka ma opór 26,45 Ω i jest podłączona do napięcia 230 V. Oblicz moc wydzieloną w grzałce.

**Rozwiązanie:**
$$
P = \frac{U^2}{R} = \frac{230^2}{26,45} \approx 2000 \text{ W}
$$

## 3. Zmiana napięcia
Żelazko 2000 W, 230 V podłączamy do sieci 200 V. Oblicz moc, jaką będzie pobierać.

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{2000} \approx 26,45 \, \Omega
$$
$$
P_{200V} = \frac{200^2}{26,45} \approx 1512 \text{ W}
$$

## 4. Zbyt wysokie napięcie
Żarówka 80 W, 230 V zostaje podłączona do sieci 250 V. Oblicz moc i oceń, co się stanie z żarówką.

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{80} \approx 661 \, \Omega
$$
$$
P_{250V} = \frac{250^2}{661} \approx 94 \text{ W}
$$

**Wniosek:** żarówka pobiera więcej mocy niż przewidziano, może się przegrzać i spalić.

## 5. Natężenie i moc
Czajnik elektryczny ma moc 2200 W przy 230 V. Oblicz natężenie prądu płynącego przez czajnik.

**Rozwiązanie:**
$$
I = \frac{P}{U} = \frac{2200}{230} \approx 9,57 \text{ A}
$$

## 6. Podsumowanie bezpieczeństwa
Podłączamy urządzenie 500 W, 230 V do sieci 120 V.  
- Czy urządzenie zadziała?  
- Jaka będzie moc pobierana?  

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{500} \approx 105,8 \, \Omega
$$
$$
P_{120V} = \frac{120^2}{105,8} \approx 136 \text{ W}
$$

**Wniosek:** urządzenie działa, ale bardzo słabo, moc jest dużo niższa niż nominalna.
