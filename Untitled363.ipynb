# Słów kilka o elektryczności cz. 1

### 1. Podstawowe pojęcia

- **Napięcie elektryczne (U)**: różnica potencjałów elektrycznych między dwoma punktami obwodu. Jednostka: **wolt (V)**  
  Przykład: napięcie w gniazdku domowym wynosi 230 V

- **Opór elektryczny (R)**: opór, jaki stawia materiał przepływowi prądu. Jednostka: **om (Ω)**  
  Przykład: opór żarówki 40 W wynosi około 1320 Ω

- **Natężenie prądu elektrycznego (I)**: ilość ładunków elektrycznych przepływających przez przekrój przewodnika w jednostce czasu. Jednostka: **amper (A)**  
  Przykład: prąd płynący przez czajnik elektryczny wynosi około 8,5 A

### 2. Wzory podstawowe

- **Prawo Ohma**:  
  $$U = I \cdot R$$
- **Obliczanie oporu**:  
  $$R = \frac{U}{I}$$
- **Obliczanie natężenia prądu**:  
  $$I = \frac{U}{R}$$

### 3. Obliczanie mocy

- **Moc elektryczna**:  
  $$P = U \cdot I$$
- **Moc z oporu**:  
  $$P = \frac{U^2}{R}$$
- **Moc z natężenia**:  
  $$P = I^2 \cdot R$$

### 4. Bezpieczeństwo użytkowania urządzeń elektrycznych

- **Dobór napięcia**: Urządzenia elektryczne są projektowane do pracy przy określonym napięciu  
  - Podłączenie do niższego napięcia → urządzenie działa słabiej lub wcale  
  - Podłączenie do wyższego napięcia → urządzenie może ulec uszkodzeniu

- **Bezpieczniki**: chronią obwody przed przeciążeniem i zwarciem.  
  W domowych instalacjach stosuje się bezpieczniki o wartościach np. 10 A lub 16 A

### 1. Prawo Ohma
Żarówka ma opór 120 Ω. Oblicz natężenie prądu płynącego przez żarówkę, jeśli podłączymy ją do sieci o napięciu 230 V.

**Rozwiązanie:**
$$
I = \frac{U}{R} = \frac{230}{120} \approx 1,92 \text{ A}
$$

### 2. Moc z oporu
Grzałka żelazka ma opór 26,45 Ω i jest podłączona do napięcia 230 V. Oblicz moc wydzieloną w grzałce.

**Rozwiązanie:**
$$
P = \frac{U^2}{R} = \frac{230^2}{26,45} \approx 2000 \text{ W}
$$

### 3. Zmiana napięcia
Żelazko 2000 W, 230 V podłączamy do sieci 200 V. Oblicz moc, jaką będzie pobierać.

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{2000} \approx 26,45 \, \Omega
$$
$$
P_{200V} = \frac{200^2}{26,45} \approx 1512 \text{ W}
$$

### 4. Zbyt wysokie napięcie
Żarówka 80 W, 230 V zostaje podłączona do sieci 250 V. Oblicz moc i oceń, co się stanie z żarówką.

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{80} \approx 661 \, \Omega
$$
$$
P_{250V} = \frac{250^2}{661} \approx 94 \text{ W}
$$

**Wniosek:** żarówka pobiera więcej mocy niż przewidziano, może się przegrzać i spalić.

### 5. Natężenie i moc
Czajnik elektryczny ma moc 2200 W przy 230 V. Oblicz natężenie prądu płynącego przez czajnik.

**Rozwiązanie:**
$$
I = \frac{P}{U} = \frac{2200}{230} \approx 9,57 \text{ A}
$$

### 6. Podsumowanie bezpieczeństwa
Podłączamy urządzenie 500 W, 230 V do sieci 120 V.  
- Czy urządzenie zadziała?  
- Jaka będzie moc pobierana?  

**Rozwiązanie:**
$$
R = \frac{U^2}{P} = \frac{230^2}{500} \approx 105,8 \, \Omega
$$
$$
P_{120V} = \frac{120^2}{105,8} \approx 136 \text{ W}
$$

**Wniosek:** urządzenie działa, ale bardzo słabo, moc jest dużo niższa niż nominalna.


# Słów kilka o elektryczności cz. 2

## 1. Obwód elektryczny

Obwód elektryczny to zamknięta droga, którą przepływa prąd elektryczny. Składa się z:

- **Źródła napięcia** (np. bateria, akumulator, gniazdko elektryczne),
- **Odbiornika** (np. żarówka, silnik),
- **Przewodów** łączących te elementy.

Aby prąd płynął, obwód musi być zamknięty — przerwanie obwodu (np. wyłączenie przełącznika) powoduje zatrzymanie przepływu prądu.

## 2. Połączenia szeregowe i równoległe

### Połączenie szeregowe

- **Napięcie**: sumuje się na poszczególnych odbiornikach.
- **Natężenie prądu**: jest takie samo w całym obwodzie.
- **Opór**: sumuje się;
$$
R_{\text{całkowity}} = R_1 + R_2 + \dots + R_n
$$

### Połączenie równoległe

- **Napięcie**: jest takie samo na wszystkich odbiornikach.
- **Natężenie prądu**: sumuje się;
$$
I_{\text{całkowity}} = I_1 + I_2 + \dots + I_n
$$
- **Opór**: odwrotności sumują się;
$$
\frac{1}{R_{\text{całkowity}}} = \frac{1}{R_1} + \frac{1}{R_2} + \dots + \frac{1}{R_n}
$$

## 3. Moc elektryczna

Moc elektryczna ($P$) to ilość energii przekazywanej lub zamienianej w jednostce czasu. Wyrażana jest wzorem:

$$
P = U \cdot I
$$

gdzie:
- $( P $) – moc w watach (W),
- $( U $) – napięcie w woltach (V),
- $( I $) – natężenie prądu w amperach (A).

Moc może być również wyrażona przez opór:

$$
P = \frac{U^2}{R} \quad \text{lub} \quad P = I^2 \cdot R
$$

## 4. Przykład obliczeniowy

Załóżmy, że mamy dwie żarówki o oporach
$R_1 = 10 \, \Omega$ i $R_2 = 20 \, \Omega$,
podłączone do napięcia $U = 230 \, \text{V}$.


### Połączenie szeregowe

- Całkowity opór:
$$
R_{\text{całkowity}} = R_1 + R_2 = 10 + 20 = 30 \, \Omega
$$
- Natężenie prądu:
$$
I = \frac{U}{R_{\text{całkowity}}} = \frac{230}{30} \approx 7,67 \, \text{A}
$$
- Moc:
$$
P = U \cdot I = 230 \cdot 7,67 \approx 1764,1 \, \text{W}
$$

### Połączenie równoległe

- Całkowity opór:
$$
\frac{1}{R_{\text{całkowity}}} = \frac{1}{R_1} + \frac{1}{R_2} = \frac{1}{10} + \frac{1}{20} = \frac{3}{20}
$$
$$
R_{\text{całkowity}} = \frac{20}{3} \approx 6,67 \, \Omega
$$
- Natężenie prądu:
$$
I = \frac{U}{R_{\text{całkowity}}} = \frac{230}{6,67} \approx 34,5 \, \text{A}
$$
- Moc:
$$
P = U \cdot I = 230 \cdot 34,5 \approx 7935 \, \text{W}
$$

## 5. Podsumowanie

- **Połączenie szeregowe**: obniża natężenie prądu, zwiększa opór, zmniejsza moc.
- **Połączenie równoległe**: zwiększa natężenie prądu, zmniejsza opór, zwiększa moc.
- **Moc elektryczna**: zależy od napięcia, natężenia prądu i oporu; w połączeniu szeregowym moc jest mniejsza niż w równoległym przy tym samym napięciu.


### Zadania z elektryczności – część 2

### Zadanie 1 – połączenie szeregowe
Dwie żarówki o oporach $R_1 = 15 \, \Omega$ i $R_2 = 25 \, \Omega$ są podłączone szeregowo do źródła napięcia $U = 230 \, \text{V}$.  
Oblicz:
1. Całkowity opór obwodu.
2. Natężenie prądu w obwodzie.
3. Moc wydzieloną w każdej żarówce i w całym obwodzie.

### Zadanie 2 – połączenie równoległe
Te same żarówki ($R_1 = 15 \, \Omega$, $R_2 = 25 \, \Omega$) podłączamy równolegle do źródła napięcia $U = 230 \, \text{V}$.  
Oblicz:
1. Całkowity opór obwodu.
2. Natężenie prądu płynącego przez każdą żarówkę i całkowite natężenie.
3. Moc wydzieloną w każdej żarówce i w całym obwodzie.

### Zadanie 3 – zmiana napięcia
Żelazko o mocy $P = 2000 \, \text{W}$ i napięciu projektowym $U = 230 \, \text{V}$ zostaje podłączone do sieci $U = 200 \, \text{V}$.  
Oblicz:
1. Opór żelazka.
2. Moc pobieraną przy napięciu 200 V.

### Zadanie 4 – zbyt wysokie napięcie
Żarówka o mocy $P = 60 \, \text{W}$ i napięciu $U = 230 \, \text{V}$ zostaje podłączona do sieci $U = 250 \, \text{V}$.  
Oblicz:
1. Opór żarówki.
2. Moc pobieraną przy 250 V.
3. Oceń, co się stanie z żarówką.

### Rozwiązania

### Zadanie 1 – połączenie szeregowe
Całkowity opór:
$$
R_{\text{całkowity}} = R_1 + R_2 = 15 + 25 = 40 \, \Omega
$$

Natężenie prądu:
$$
I = \frac{U}{R_{\text{całkowity}}} = \frac{230}{40} \approx 5,75 \, \text{A}
$$

Moc w każdej żarówce:
$$
P_1 = I^2 \cdot R_1 = 5,75^2 \cdot 15 \approx 496 \, \text{W}
$$
$$
P_2 = I^2 \cdot R_2 = 5,75^2 \cdot 25 \approx 826 \, \text{W}
$$

Moc całkowita:
$$
P_{\text{całkowita}} = P_1 + P_2 \approx 1322 \, \text{W}
$$

### Zadanie 2 – połączenie równoległe
Całkowity opór:
$$
\frac{1}{R_{\text{całkowity}}} = \frac{1}{R_1} + \frac{1}{R_2} = \frac{1}{15} + \frac{1}{25} = \frac{8}{75}
$$
$$
R_{\text{całkowity}} = \frac{75}{8} \approx 9,375 \, \Omega
$$

Natężenie prądu przez każdą żarówkę:
$$
I_1 = \frac{U}{R_1} = \frac{230}{15} \approx 15,33 \, \text{A}
$$
$$
I_2 = \frac{U}{R_2} = \frac{230}{25} \approx 9,2 \, \text{A}
$$
$$
I_{\text{całkowite}} = I_1 + I_2 \approx 24,53 \, \text{A}
$$

Moc w każdej żarówce:
$$
P_1 = U \cdot I_1 = 230 \cdot 15,33 \approx 3526 \, \text{W}
$$
$$
P_2 = U \cdot I_2 = 230 \cdot 9,2 \approx 2116 \, \text{W}
$$
$$
P_{\text{całkowita}} = P_1 + P_2 \approx 5642 \, \text{W}
$$

### Zadanie 3 – zmiana napięcia
Opór żelazka:
$$
R = \frac{U^2}{P} = \frac{230^2}{2000} \approx 26,45 \, \Omega
$$

Moc przy $U = 200 \, \text{V}$:
$$
P_{200V} = \frac{U^2}{R} = \frac{200^2}{26,45} \approx 1512 \, \text{W}
$$

### Zadanie 4 – zbyt wysokie napięcie
Opór żarówki:
$$
R = \frac{U^2}{P} = \frac{230^2}{60} \approx 881,67 \, \Omega
$$

Moc przy $U = 250 \, \text{V}$:
$$
P_{250V} = \frac{250^2}{881,67} \approx 70,8 \, \text{W}
$$

**Wniosek:** żarówka pobiera więcej mocy niż przewidziano, może się przegrzać i spalić.
